```python
delivery = pd.read_csv('deliveries.csv')
delivery.head(2)
```

```python
batsmans['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
# This is wrong method
batsmans.get_group('V Kohli').sort_values('batsman_runs', ascending=False)[['bowling_team', 'batsman_runs']].head(3)
```

```python
vk = delivery[delivery['batsman'] == 'V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def runScored(batsman_name):
    vk = delivery[delivery['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
print(runScored('V Kohli'))
print(runScored('G Gambhir'))
print(runScored('MS Dhoni'))
```
